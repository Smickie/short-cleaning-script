```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
from gspread_dataframe import get_as_dataframe, set_with_dataframe
import os

path = os.getcwd()

scope = ['https://spreadsheets.google.com/feeds',
         'https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name(path + '/keys/cision-418313-bb0e379d7fe8.json', scope)
client = gspread.authorize(creds)

sheet = client.open_by_url('https://docs.google.com/spreadsheets/d/1qkdcgKvlvaaJeoEstA22Bu5a_1IjHJWEYvT-tSDJbII/edit#gid=0').sheet1

df = get_as_dataframe(sheet, header=None)

df = df[[0,1]]

df = df.dropna()

df.columns = ["t", "a"]



tdmask = df["t"].str.match(r"^(?!\d{4}-\d{2}-\d{2}).*\d{2}:\d{2}(:\d{2})?")
ymask = df["t"].str.match(r"^(\d{4}-\d{2}-\d{2}).*\d{2}:\d{2}(:\d{2})?")

df1 = df.loc[tdmask]
df2 = df.loc[ymask]

df1["t"] = df1["t"].str.extract(r"(\d+[\d:]\d+)")

df2["t"] = df2["t"].str.extract(r"(\d{4}-\d{1,2}-\d{1,2})")

df2['t'] = df2['t'].replace('', pd.NA).ffill().where(df2['t'].ne(df2['t'].shift()), '')

ndf = pd.concat([df1, df2]).reset_index(drop=True)

english_keywords = ['Annual', 'Notice', 'presents', "resolutions", "revised", "Meeting", "MEETING", "ANNUAL",
                   "appoints", "Appoints", "APPOINTS", "new", "notice", "NOTICE", "Notice", "additional", 
                   "establishing", "Interim", "interim", "acquires", "strengthens", "protection", "extends", "secures",
                   "change", "owner", "dividend", "signed", "rebrands", "year-end", "collaborate", "receives"]

rows_to_remove = df['a'].str.contains('|'.join(english_keywords), case=False, na=False)

ndf = ndf[~rows_to_remove].reset_index(drop=True)

ndf.head(60)
```

```python
sheet.clear()
set_with_dataframe(sheet, ndf, include_index=False, include_column_header=False)
```
